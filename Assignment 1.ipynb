```python
# Question 1

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
duplicate_transaction = []
fraud_transaction = []
total_amount = 0
unique_transaction = []

for i in range(len(transactions)):
    for j in range(i + 1, len(transactions)):
        if transactions[i] == transactions[j]:
            duplicate_transaction.append(transactions[i])
    if transactions[i] > 5000:
        fraud_transaction.append(transactions[i])
    total_amount = total_amount + transactions[i]

unique_transaction = [x for x in transactions if x not in duplicate_transaction]
print('All transaction values that appear more than once')
print(duplicate_transaction)
print('Detect transactions greater than ₹5000 as these may require fraud monitoring')                                 
print(fraud_transaction)
print('Total spending for the day')
print(total_amount)
print('Unique transaction')
print(unique_transaction)
unique_transaction.sort(reverse = True)
print('Sort the unique transaction list in descending order to view the largest transactions first.')
print(unique_transaction)
```

```python
# Question 4

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_login_hours = []
odd_login_hours = []
peak_login_hours = []

for i in range(len(login_hours)):
    if (login_hours[i] % 2) != 1:
        even_login_hours.append(login_hours[i])
    else:
        odd_login_hours.append(login_hours[i])
    if login_hours[i] > 5:
        peak_login_hours.append(login_hours[i])
print(even_login_hours)
print(odd_login_hours)
print(peak_login_hours)

login_hours.sort()
print(login_hours)

print(len(peak_login_hours))
```
