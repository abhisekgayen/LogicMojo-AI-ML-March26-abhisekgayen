```python
#Q1. Division with Exception Handling
#Write a program that takes two numbers and safely performs division. Handle division by zero using try-except.

a = int(input('Enter first Number'))
b = int(input('Enter second number'))

try:
    print(a/b)
except ZeroDivisionError:
    print("Second number can not be zero")
    
```

```python
#Q2. String to Integer Conversion
#Given a list: data = ["10", "20", "abc", "30"]
#Convert all values to integers. Skip invalid values using exception handling.

data = ["10", "20", "abc", "30"]
result = []

for item in data:
    try:
        num = int(item)
        result.append(num)
    except ValueError:
        # Skip invalid values
        pass

print(result)
```

```python
#Q3. Multiple Exception Handling
#Write a program that: Takes user input Converts it to integer Divides 100 by that number
#Handle: ValueError and ZeroDivisionError

def multiple_Exception_Handling():
    num = input('Provide your input')
    try:
        num1 = int(num)
        print(100/num1)
    except ValueError:
        print('Enter value not conveted to integer')
    except ZeroDivisionError:
        print('Enter value can not be zero')
multiple_Exception_Handling()
```

```python
#Q4. Using finally Write a program that: 
#Opens a file (dummy) Prints a message Ensures a “Closing file” message always runs using finally

file = open("data.csv")

try:
    data = file.read()
except:
    print('Error reading the file')
finally:
    file.close()
    print('Closing file')
```

```python
#Q6. Clean Text
#Given: text = " Hello Python "
#Remove spaces and convert to lowercase.

import re
text = " Hello Python "

result = re.sub("^\s+|\s$", "", text).lower()
print(result)

result2 = re.sub("^\s+|\s$|\s", "", text).lower() # Reoved all specess

print(result2)
```

```python
#Q7. Replace Words Replace all occurrences of "Python" with "AI" in:
#text = "Python is great. Python is easy."

import re
text = "Python is great. Python is easy."

result = re.sub("Python", "AI", text)
print(result)
```

```python
#Q8. Split and Join
#Convert: text = "apple,banana,orange" Into: "apple banana orange"

import re
text = "apple,banana,orange"

result = re.split(",", text)
result1 = " ".join(result)
print(result1)
```

```python
#Q9. Count Words 
#Count number of words in a sentence using split.

import re
text = "My Name is Abhisek Gayen. I am learning AI/ML. I live in kolkata"
word_list = re.split(" ", text)
print(len(word_list))
```

```python
#Q10. Reverse Words
#Reverse words in a string: "I love AI"

import re

text = "I love AI"
result = " ".join(text.split()[::-1])
print(result)
```

```python
#Q11. Find Digits
#Extract all digits from: text = "My number is 98765"

import re
text = "My number is 98765"
result = re.search(r"\d+", text)
print(result.group())
```

```python
#Q12. Find Words Starting with 'a'
#From: text = "apple banana avocado berry"
#Extract words starting with 'a'.

import re
text = "apple banana avocado berry"
result = re.findall(r"\ba\w*", text)
print(result)
```

```python
#Q13. Extract Emails
#Extract emails from: text = "Contact: test@gmail.com, admin@yahoo.com"

import re
text = "Contact: test@gmail.com, admin@yahoo.com"

email = re.findall(r"\S+@\S+", text)

print(email)
```

```python
#Q14. Validate Phone Numbers
#Extract valid 10-digit phone numbers from: text = "Call 9876543210 or 1234567890"

import re
text = "Call 9876543210 or 12345"
result = re.findall("\b\d{10}\b", text)
print(result)
```
