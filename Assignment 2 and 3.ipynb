Write a function detect_increasing_sequence(data)
Find the longest continuous increasing sequence
Return that sequence as a list
Example output:
[10, 20, 30]

```python
def detect_increasing_sequence(data):
    longest = []
    current = [data[0]]

    for i in range(1, len(data)):
        if data[i] > data[i - 1]:
            current.append(data[i])
        else:
            if len(current) > len(longest):
                longest = current
            current = [data[i]]

    if len(current) > len(longest):
        longest = current

    return longest
data = [10, 20, 30, 25, 40, 50, 60, 70, 80]

print(detect_increasing_sequence(data))
```

You are given:

user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
Task:

Write a function recommend(user)
Recommend items based on common items between users
Do not recommend items already purchased by the user
Use sets and dictionary iteration

```python
def recommend(user, user_purchases):
    target_items = set(user_purchases[user])
    recommendations = set()

    for other_user, items in user_purchases.items():
        if other_user != user:
            other_items = set(items)
            if target_items & other_items:
                recommendations |= other_items
                
    recommendations -= target_items

    return list(recommendations)

user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

print(recommend("user1", user_purchases))
```

Task:

Write a recursive function decision_tree(score)
If score is greater than 80 return "Approved"
If score is between 50 and 80 call the function again with score plus 10
If score is less than 50 return "Rejected"

```python
score = int(input('Please enter your score'))

def decision_tree(score):
    if score >= 80:
        print('Approved')
    elif score > 50 and score < 80:
        score = score + 10
        decision_tree(score)
    else:
        print('Rejected')

decision_tree(score)
```
