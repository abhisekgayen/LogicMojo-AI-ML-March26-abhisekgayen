```python
#Q1. Understanding Feature Distributions (Iris Dataset)
#Context: Before training any ML model, we analyze feature distributions.
#Dataset: Load Iris dataset from sklearn
#Tasks:
#Extract feature matrix as NumPy array
#Compute: Mean Median Standard deviation Variance (for each feature)
#Identify: Which feature has highest variability and why it matters in ML
#Convert any one feature into shape (n,1) and explain why ML models expect this format
######################################################################################

import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
iris = load_iris()

feature_matrix = iris.data
feature_names = iris.feature_names

df = pd.DataFrame(feature_matrix, columns=iris.feature_names )

#Mean Median Standard deviation Variance (for each feature)
#df.info()
print(df.describe())
#Which feature has highest variability, this one needed for better separation between classes
print(df.var())

#Convert any one feature into shape (n,1) and why ML models expect this format beacuse input format need as 2D array and output as a 1D Array
sepal_length = np.array(df['sepal length (cm)'])
print(sepal_length.reshape(-1,1))
```

```python
#Q2. Data Selection for Model Input
#Context: Models are trained on selective features, not full raw data.
#Dataset: Iris
#Tasks:
#1. Extract: First 100 samples Only last 2 features
#2. Use boolean masking: Select samples where petal length is greater than dataset mean
#3. Count selected samples
#4. Explain how this relates to feature-based filtering in ML
#########################################################################

#Extract: First 100 samples Only last 2 features
print(df.iloc[0:100, 2:4])

mask = df['petal length (cm)'] > df['petal length (cm)'].mean()
print(df[mask])

#Count selected samples
print(len(df[mask]))

# Boolean masking need for clear unwanted data and focus on specific group data
```

```python
#Q3. Feature Scaling using NumPy (Very Important)
#Context: Most ML models require normalized data.
#Dataset: Iris (NumPy array)
#Tasks:
#1. Apply standardization
#2. Verify: Mean ≈ 0 Std ≈ on

#formula [new std value = (original value - mean of a feature)/ feature standard division]

new_value = (feature_matrix - np.mean(feature_matrix)) / np.std(feature_matrix)

print('Mean:', np.mean(new_value))
print('Std:', np.std(new_value))
```

```python
#Q4. Dataset Understanding (Titanic Dataset)
#Context: First step in ML pipeline is dataset inspection.
#Dataset: Titanic (Seaborn)
#Tasks:
#Display: head(), tail() info(), describe()
#Identify: Missing values per column Numerical vs categorical features
#Explain: Why identifying feature types is important before modeling
####################################################################

import numpy as np
import pandas as pd
import seaborn as sns

data = sns.load_dataset("titanic")

titanic_df = data.copy()

#Display: head(), tail() info(), describe()
titanic_df.head()
titanic_df.tail()
titanic_df.info()

#Identify: Missing values per column Numerical vs categorical features
titanic_df.isnull().sum()

# we need to identify missing value for each features because we need to fill missing value with mean, midian or mode value so our ML model can predict
# correct output
```

```python
#Q5. Filtering Data for Business Logic
#Context: Real ML problems require filtering relevant populations.
#Dataset: Titanic
#Tasks:
#1. Filter: Female passengers in 1st class
#2. Compute: Survival rate for this group
#3. Compare with overall survival rate
#4. Interpret: What insight can be used in ML feature engineering
#######################################################################

titanic_df.head()

#1. Filter: Female passengers in 1st class
female_passenger = (titanic_df['sex'] == 'female') & (titanic_df['class'].isin(['First']))
female_survive_first_class = (titanic_df['sex'] == 'female') & (titanic_df['class'].isin(['First']) & (titanic_df['survived'] == 1))
titanic_df[female_passenger]
titanic_df[female_survive_first_class]

#2. Compute: Survival rate for this group
survival_rate = (len(titanic_df[female_survive_first_class])/len(titanic_df[female_passenger])) * 100
print(survival_rate)

# This feature can provide high survival rate group
```

```python
#Q6. Handling Missing Data (Critical ML Step)
#Context: Models cannot handle null values.
#Dataset: Titanic
#Tasks:
#1.Fill missing age using median
#2.Fill missing embarked using mode
#3.Drop deck column
#4.Explain: Why different strategies are used for different columns
####################################################################

#1.Fill missing age using median
titanic_df['age'].fillna(titanic_df['age'].median(), inplace = True)

#2.Fill missing embarked using mode
titanic_df['embarked'].fillna(titanic_df['embarked'].mode()[0], inplace = True)

#3.Drop deck column
titanic_df.drop(columns = ['deck'], inplace = True)
titanic_df.isnull().sum()
```

```python
#Q7. Feature Engineering for Model Improvement
#Context: Raw data is rarely useful—features must be created.
#Dataset: Titanic
#Tasks:
#Create: family_size = sibsp + parch + 1
#Create: is_alone (binary feature)
#Compute survival rate by is_alone
#Explain: Why engineered features improve ML models
###########################################################

#Create: family_size = sibsp + parch + 1
titanic_df['family_size'] = titanic_df['sibsp'] + titanic_df['parch'] + 1

#Create: is_alone (binary feature)
titanic_df['is_alone'] = titanic_df['family_size']<= 1

#Compute survival rate by is_alone

alone_data = titanic_df['is_alone']
alone_survived = (titanic_df['is_alone']) & (titanic_df['survived'])
titanic_df[alone_data]
titanic_df[alone_survived]
survival_rate_alone = (len(titanic_df[alone_survived]) / len(titanic_df[alone_data])) * 100
print(survival_rate_alone)
```

```python
#Q8. Group-Based Insights (Customer Segmentation Logic)
#Context: ML models often depend on aggregated patterns.
#Dataset: Tips
#Tasks:
#Compute: Average tip by day Total bill by gender
#Create new feature: tip_percentage
#Identify: Which day has highest tipping behavior
#Explain: How this helps in segmentation models
###################################################

import pandas as pd
import seaborn as sns

df = sns.load_dataset('tips')
data_tips = df.copy()

#Compute: Average tip by day Total bill by gender
avg_tip_day = data_tips.groupby('day')['tip'].mean()
print('Average tip by day:\n', avg_tip_day)

total_bill_gender = data_tips.groupby('sex')['total_bill'].sum()
print('Total bill by gender:\n', total_bill_gender)

#Create new feature: tip_percentage
data_tips['tip_percentage'] = (data_tips['tip'] / data_tips['total_bill']) * 100
print(data_tips.head())

#Identify: Which day has highest tipping behavior
highest_tip_day = data_tips.groupby('day')['tip'].max()
print('highest tipping behavior:\n', highest_tip_day)

# By creating features like tip_percentage and analyzing behavior across categories (day, gender), we can uncover patterns that help segment 
#customers and improve machine learning models by providing more meaningful and normalized features.
```

```python
#Q9. Advanced Filtering + Subsetting
#Context: Training data is often filtered based on business rules.
#Dataset: Tips
#Tasks:
#Filter: total_bill > 20 AND tip < 3
#Select only relevant columns
#Analyze: Is this segment under-tipping?
#Explain: How such filtering helps anomaly detection
###############################################################

data_tips.head()

#Filter: total_bill > 20 AND tip < 3
filter_data = (data_tips['total_bill'] > 20) & (data_tips['tip'] < 3)
data_tips[filter_data]

#Select only relevant columns
data_tips.loc[filter_data, ['total_bill', 'sex', 'tip', 'day', 'tip_percentage']]

#Analyze: Is this segment under-tipping?

avg_tip_percentange = data_tips['tip_percentage'].mean()
avg_filter_data_tip_percentage = data_tips.loc[filter_data, ['tip_percentage']].mean()

print('overall tip %\n', avg_tip_percentange)
print('segment tip %\n', avg_filter_data_tip_percentage)

under_tipping = False

if avg_tip_percentange > avg_filter_data_tip_percentage['tip_percentage']:
    under_tipping = True
print(under_tipping)
```

```python
#Q11. End-to-End ML Preprocessing Pipeline (Titanic)
#Context: Prepare dataset for ML model training
#Tasks:
#Load dataset
#Handle missing values
#Encode categorical variables (sex, embarked)
#Normalize numerical features
#Convert to NumPy array
#Output: Final feature matrix ready for model training
############################################################

import pandas as pd 
import seaborn as sns
import numpy as np

df = sns.load_dataset("titanic")

titanic_data = df.copy()

titanic_data.head()
#Handle missing values
titanic_data.fillna({
    'age': titanic_data['age'].mean(),
    'embark_town': titanic_data['embark_town'].mode()[0],
    'embarked': titanic_data['embarked'].mode()[0]
}, inplace=True)

titanic_data.drop(columns = ['deck'], inplace = True)

#Encode categorical variables (sex, embarked)
titanic_data = pd.get_dummies(df, columns=['sex', 'embarked'], drop_first=True)

#Normalize numerical features
num_cols = titanic_data.select_dtypes(include=['int64', 'float64']).columns # Indetified numeric columns
for col in num_cols:
    titanic_data[col] = (titanic_data[col] - titanic_data[col].min()) / (titanic_data[col].max() - titanic_data[col].min())

#Convert to NumPy array
titanic_data_arr = np.array(titanic_data)
#Output: Final feature matrix ready for model training
print(titanic_data_arr)
```

```python
#Q12. Feature Selection using Statistical Understanding
#Context: Not all features improve model performance.
#Dataset: Iris
#Tasks:
#Compute correlation matrix
#Identify highly correlated features
#Drop redundant features
#Explain: Impact on overfitting and model performance
#########################################################

import pandas as pd 
import seaborn as sns

df = sns.load_dataset("iris")
iris_dataset = df.copy()

#Compute correlation matrix
corr_matrix = iris_dataset.select_dtypes(include=['int64', 'float64']).corr()

#Identify highly correlated features
threshold = 0.9

high_corr_feature = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            high_corr_feature.append((col1, col2, corr_matrix.iloc[i, j]))

print(high_corr_feature)

#Drop redundant features
to_drop = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname = corr_matrix.columns[i]
            to_drop.add(colname)

df_reduced = df.drop(columns=to_drop)

print(to_drop)
```

```python
#Q13. Custom Feature Transformation using apply()
#Context: Real ML requires custom logic transformation.
#Dataset: Titanic
#Tasks:
#Create age_group: Child / Adult / Senior
#Use .apply()
#Compute survival rate per group
#Interpret: Which segment has highest survival likelihood
##################################################################

import pandas as pd 
import seaborn as sns

df = sns.load_dataset("titanic")

titanic_dataset = df.copy()

#Create age_group: Child / Adult / Senior
#Use .apply()
titanic_dataset['age_group'] = titanic_dataset['age'].apply(lambda x: 'Child' if x < 18 else ('Adult' if x < 60 else 'Senior'))

#Compute survival rate per group
age_group = (titanic_dataset.groupby('age_group')['survived'].mean()) * 100
print(age_group)
```

```python
#Q14. Error Analysis Simulation (Model Evaluation Thinking)
#Context: Understanding prediction errors is key in ML.
#Tasks:
#Generate 1000 random error values using NumPy
#Compute: Mean Standard deviation
#Identify outliers: Using mean ± 2*std
#Explain: What high variance means in model performance
###############################################################

import numpy as np
#Generate 1000 random error values using NumPy
errors = np.random.normal(loc=0, scale=1, size=1000)

#Compute: Mean Standard deviation
mean_value = np.mean(errors)
std_value = np.std(errors)
print('Mean Value:', mean_value)
print('Standard deviation', std_value)

#Identify outliers: Using mean ± 2*std
upper_limit = (mean_value + 2) * std_value
lower_limit = (mean_value - 2) * std_value
print('Using mean ± 2*std\n', upper_limit, lower_limit)
```
